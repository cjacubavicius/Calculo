# **Integral indefinida**

```python
from sympy import *
```

```python
x = Symbol('x')
```

```python
integrate(x**2)
```

```python
integrate(3*x**2+1)
```

```python
integrate(x**(1/2))
```

```python
integrate(1/x)
```

```python
integrate(x**2*exp(x**3))
```

```python
integrate(sin(x))
```

```python
integrate((sin(x))**4*cos(x))
```

```python
integrate(x*cos(x))
```

# **Integral Definida**

```python
integrate(2*x+1,(x,1,3))
```

```python
integrate(x**3-4*x**2+1,(x,0,1))
```

```python
integrate(3*x**2+2*x,(x,0,3))
```

```python
integrate(6/(0.07*x+5))
```

```python
integrate(-0.02*(x-7)*(x-14),(x,8,20))
```
