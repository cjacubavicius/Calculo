# **EQUAÇÕES**

## **EQUAÇÕES POLINOMIAIS**

```python
import sympy as sy
# Declaração da variável
x = sy.Symbol("x")
```

```python
sy.solve(3*x+1)
```

```python
sy.solve(x**2-5*x)
```

```python
sy.solve(3*x**2-48)
```

```python
sy.solve(x**2-x-6)
```

```python
sy.solve((x-5)*(x+5)+7)
```

```python
sy.solve(x**2/3-(x-9)/6-3/2)
```
