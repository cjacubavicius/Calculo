## **FUNÇÕES POLINOMIAIS**

### **Função de 2o grau**

```python
sy.plot(x**2-2*x-3, (x, -3,5));
```

```python
sy.plot(-x**2-4, (x, -4,4));
```

```python
sy.plot(-x**2+2*x-2, (x, -4,6));
```

### **Função de Grau Superior**

```python
sy.plot(3*x**3-2*x**2-4*x-5, (x,-2,2));
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
  return x**3 + 1
```

```python
x = np.linspace(-1,1)
plt.plot(x, f(x))
plt.grid()
plt.show()
```

```python
def f(x):
  return 3*x**3-2*x**2-4*x-5

x = np.linspace(-2,2)
plt.plot(x, f(x));
```
