# Limites

## Cálculo de Limites

```python
import sympy as sy
x = sy.Symbol("x")
```

```python
lim = sy.limit(x**2 + 1, x, 1)
lim
```

```python
lim2 = sy.limit(x**2+3*x+5, x, 2)
lim2
```

```python
lim3 = sy.limit((x-5)/(x**3-7), x, 3)
lim3
```

```python
lim4 = sy.limit(sy.sqrt(x**4-4*x+1), x, -2)
lim4
```

```python
lim5 = sy.limit((x**2-1)/(x-1), x, 1)
lim5
```

## Limite Lateral

```python
lim5 = sy.limit((2*x+1)/(x-1), x, 1, dir='+')
lim5
```

```python
lim6 = sy.limit((2*x+1)/(x-1), x, 1, dir='-')
lim6
```

```python
lim7 = sy.limit((2*x+1)/(x-1), x, 1, dir='+-')
lim7
```

## Limites com fatoração

```python
lim8 = sy.limit((x**2+5*x)/(x+5), x, -5)
lim8
```

```python
lim9 = sy.limit((x**2-10*x+25)/(x-5), x, 5)
lim9
```

```python
lim10 = sy.limit((x**2-5*x+6)/(x-2), x, 2)
lim10
```

## Limite no infinito

```python
lim9 = sy.limit(x**3-x**2-4*x+4, x, sy.oo)
lim9
```

```python
lim10 = sy.limit((3/5)**x, x, sy.oo)
lim10
```

```python
lim10 = sy.limit((5/3)**(-x), x, sy.oo)
lim10
```
