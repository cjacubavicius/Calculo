# **Expressão Algébrica**

**Importação da biblioteca**

```python
import sympy as sy

# Declaração da variável
x = sy.Symbol("x")
```

## Soma

```python
(x**3 + 2*x**2 + 3*x) + (4*x**3-x**2+5*x)
```

## Expansão e Produtos Notáveis

```python
# Declaração de outras variáveis
y = sy.Symbol("y")
z = sy.Symbol("z")
```

```python
(x+2*y)*(6-2*z)
```

```python
expr = (x + 2*y)*(6 - 2*z)
expansao = sy.expand(expr)
expansao
```

```python
# Produto da soma pela diferença
expr2=(x+y)*(x-y)
expansao2 = sy.expand(expr2)
expansao2
```

```python
# Quadrado da soma
expr3 = (x+y)**2
expansao3 = sy.expand(expr3)
expansao3
```

```python
# Quadrado da diferença
expr4 = (x-y)**2
expansao4 = sy.expand(expr4)
expansao4
```

## Fatoração

```python
expr5 = x**2 - 4
fatoracao = sy.factor(expr5)
fatoracao
```

```python
expr6 = 4*x**2 + 4*x + 1
fatoracao2 = sy.factor(expr6)
fatoracao2
```

```python
expr7 = 8*x**3 + 4*x**2 -12*x
fatoracao3 = sy.factor(expr7)
fatoracao3
```

## Divisão

```python
sy.div(x**3 + x**2 - 2*x, x**2 + 2*x)
```

```python
expr7 = (x**3 + x**2 - 2*x)
quociente, resto = sy.div(expr7, x**2 + 2*x)
print("O Quociente é:", quociente)
print("o resto é:", resto)
```

```python
expr7 = (x**3 - 1)
quociente, resto = sy.div(expr7, x - 1)
print("Quociente:", quociente)
print("Resto:", resto)
```
