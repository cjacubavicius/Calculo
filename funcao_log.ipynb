# **FUNÇÕES**

```python
import sympy as sy
```

```python
x = sy.Symbol('x')
```

```python
from sympy import plotting
```

# **FUNÇÕES LOGARÍTMICA**

```python
sy.plot(sy.log(x, 1/4), (x,-1,6));
```

```python
sy.plot(sy.log(x, 2), (x,-1,8));
```

```python
sy.plot(sy.log(x, 1/2), (x,-2,8));
```

### **Função Modular**

```python
sy.plot(abs(2*x),(x,-2,2));
```

```python
sy.plot(abs(3*x-2), (x,-2,3));
```

```python
sy.plot(x*abs(x), (x,-2,2));
```
