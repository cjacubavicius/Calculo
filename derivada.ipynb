# **DERIVADA**

## **Derivadas Fundamentais**

```python
import sympy as sy
x = sy.Symbol("x")
```

```python
der1 = sy.diff(x**2, x)
der1
```

```python
der2 = sy.diff(x**3 + sy.cos(x) - sy.sin(x) -4, x)
der2
```

```python
der3 = sy.diff(sy.tan(x)+sy.exp(x))
der3
```

```python
der4 = sy.diff(x**2 * sy.exp(x), x)
der4
```

```python
der5 = sy.diff(x**3 / sy.cos(x))
der5
```
