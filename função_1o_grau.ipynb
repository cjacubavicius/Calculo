### 1o Grau

```python
import sympy as sy
```

```python
x = sy.Symbol('x')
```

```python
from sympy import plotting
```

```python
sy.plot(3*x+1);
```

```python
sy.plot(3*x+1, (x, -1,1), title = 'Função de primeiro grau', legend = True);
```

```python
sy.plot(-4*x+3, (x, -3,3));
```

```python
sy.plot(x,(x,-2,2));
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def funcao(x):
  return 2*x+4
```

```python
x = np.linspace(-3,3)
plt.plot(x, funcao(x))
plt.grid()
plt.show()
```
