# **FUNÇÕES**

```python
import sympy as sy
```

```python
x = sy.Symbol('x')
```

```python
from sympy import plotting
```

# **FUNÇÕES TRIGONOMÉTRICAS**

```python
sy.plot(sy.sin(x));
```

```python
sy.plot(3*sy.sin(x));
```

```python
sy.plot(sy.cos(x-sy.pi/2));
```

```python
sy.plot(2*sy.cos(x));
```

```python
sy.plot(sy.sin(x+3*sy.pi/2));
```

```python
sy.plot(sy.tan(x), (x,0,10));
```

```python
sy.plot(100-20*sy.cos((8*3.14/3)*x), (x,0,2));
```
